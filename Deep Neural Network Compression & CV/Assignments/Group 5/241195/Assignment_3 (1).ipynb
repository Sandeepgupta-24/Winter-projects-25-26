1. Implement a Gaussian kernel generator:
The function signature should be <code>gaussian_kernel(ksize, sigma)</code>
It should return a numpy matrix that is the normalsied gaussian kernel with dimensions ksize x ksize and the 2D gaussian distribution calculation done with standard deviation as sigma.
````python
def convolve(img, kernel):
    # manual convolution implementation using NumPy
    i_h, i_w = img.shape[:2]
    k_h, k_w = kernel.shape[:2]
    pad = k_h // 2

    # reflect padding helps maintain edge details without dark borders
    padded = np.pad(img, pad, mode='reflect')
    output = np.zeros_like(img)

    # sliding the window manually
    for y in range(i_h):
        for x in range(i_w):
            roi = padded[y : y + k_h, x : x + k_w]
            output[y, x] = np.sum(roi * kernel)

    return output

print("manual convolution engine is initialized.")
````

````python
def gaussian_kernel(ksize, sigma):
    # 2D gaussian distribution calculation done with sigma
    ax = np.linspace(-(ksize // 2), ksize // 2, ksize)
    xx, yy = np.meshgrid(ax, ax)
    kernel = np.exp(-(xx**2 + yy**2) / (2. * sigma**2))
    return kernel / np.sum(kernel) # normalizeing the kernel

def blur_explorer(mode, k_size, sigma):
    if mode == 'Average':
        kernel = np.ones((k_size, k_size)) / (k_size**2)
    else:
        kernel = gaussian_kernel(k_size, sigma)

    blurred = convolve(img_gray, kernel)

    plt.figure(figsize=(10, 5))
    plt.imshow(blurred, cmap='gray')
    plt.title(f"{mode} Blur ;Kernel: {k_size}x{k_size}")
    plt.axis('off')
    plt.show()

interact(blur_explorer, mode=['Gaussian', 'Average'], k_size=[3, 5, 7, 9], sigma=(0.1, 5.0))
````

### Part B

#### Task - Sobel Edge Detection (NumPy Only)
Implement Sobel X and Sobel Y filters manually using convolution. Apply on the grayscale image.

Compute:
1. Gx
2. Gy
3. Gradient magnitude

Normalize to 0255 and save the edge map. Threshold if you wish to, if you do then make the thresholding part clear using code comments.

````python
# Manual Sobel Filters
sx_k = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sy_k = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

def run_sobel(threshold):
    # computing Gx and Gy
    Gx = convolve(img_gray, sx_k)
    Gy = convolve(img_gray, sy_k)

    # grad. magnitude calculation
    mag = np.sqrt(Gx**2 + Gy**2)
    mag_norm = (mag / mag.max()) * 255 # Normalize to 0-255

    # applying threshold
    edges = np.where(mag_norm > threshold, 255, 0)

    plt.imshow(edges, cmap='gray')
    plt.title(f"Sobel Edges (Threshold={threshold})")
    plt.show()

interact(run_sobel, threshold=(10, 150))
````

````python
def sharpening_tool(mode, alpha):
    if mode == 'Laplacian':
        # Use 3x3 laplacian 8 connected kernel
        lap_k = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
        lap_map = convolve(img_gray, lap_k)
        # sharp = img - lap * alpha (kernel sign affects add/sub)
        res = np.clip(img_gray + alpha * lap_map, 0, 255)
    else:
        # Unsharp Masking: sharp = img + alpha * mask where mask = img - blurred
        blurred = convolve(img_gray, np.ones((5,5))/25)
        mask = img_gray - blurred
        res = np.clip(img_gray + alpha * mask, 0, 255)

    plt.imshow(res, cmap='gray')
    plt.title(f"{mode} Sharpening (Alpha={alpha})")
    plt.show()

# Verification,,, Put alpha=1 for Laplacian and compare with cv2
interact(sharpening_tool, mode=['Laplacian', 'Unsharp'], alpha=(0.0, 2.0))
````

````python

#"Hue rotation preserves texture, only color identity changes"

def perfected_color_swap(h_min, h_max, s_min, v_factor):
    # 1. Convert RGB to HSV
    # We use cv2.cvtColor as permitted for this bonus task
    hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV).astype(np.float32)

    # 2. DESIGNING THE PERFECT MASK
    # We use a lower h_min (around 18) to catch orange transition edges
    # we  add a saturation (S) floor to avoid shifting gray/background noise
    lower_yellow = np.array([h_min, s_min, 50])
    upper_yellow = np.array([h_max, 255, 255])

    # Create the binary mask
    mask = cv2.inRange(hsv, lower_yellow, upper_yellow)

    # 3.HUE ROTATION
    # Red is 0 on the color wheel. We shift the yellow pixels to the red identity.
    hsv[mask > 0, 0] = 0

    # 4. APPLYING BRIGHTNESS FACTOR
    # To get the "less brighter colour" result, we scale the Value (V) channel.
    # Mathematically: new = old * v_factor
    hsv[mask > 0, 2] = np.clip(hsv[mask > 0, 2] * v_factor, 0, 255)

    # 5. SATURATION BOOST (Optional Vibrance)
    # To make the red "pop" like the reference, we can slightly boost saturation
    hsv[mask > 0, 1] = np.clip(hsv[mask > 0, 1] * 1.1, 0, 255)

    # Convert back to RGB for output
    final_res = cv2.cvtColor(hsv.astype(np.uint8), cv2.COLOR_HSV2RGB)

    plt.figure(figsize=(12, 6))
    plt.imshow(final_res)
    plt.title(f"Red Leaf (H-Range: {h_min}-{h_max}, Brightness: {v_factor})")
    plt.axis('off')
    plt.show()

# INTERACTIVE tuning
interact(perfected_color_swap,
         h_min=IntSlider(min=10, max=30, value=17, description="Hue Min"),
         h_max=IntSlider(min=30, max=50, value=38, description="Hue Max"),
         s_min=IntSlider(min=0, max=100, value=70, description="Sat Floor"),
         v_factor=FloatSlider(min=0.5, max=1.2, step=0.05, value=0.85, description="Brightness"))
````
